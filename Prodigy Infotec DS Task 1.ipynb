Importing important libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Total_pop = pd.read_csv('F:\Programing assignments\Prodigy_infotech DS intenship\Population Task 1\Population_data.csv')
Male_pop = pd.read_csv('F:\Programing assignments\Prodigy_infotech DS intenship\Population Task 1\Male Population.csv')
Female_pop = pd.read_csv('F:\\Programing assignments\\Prodigy_infotech DS intenship\\Population Task 1\\Female Population.csv')
```

```python
Total_pop = Total_pop.drop(columns=Total_pop.columns[1:4])
Male_pop = Male_pop.drop(columns=Male_pop.columns[1:4])
Female_pop = Female_pop.drop(columns=Female_pop.columns[1:4])
```

```python
print("Unique Countries in Total Population Dataset:", Total_pop['Country Name'].unique())
```

### Comparison of Top 20 countries by Total population in 1960, 2000 and 2023

```python
top_total_1960 = Total_pop.nlargest(20, '1960')[['Country Name', '1960']]
top_total_2000 = Total_pop.nlargest(20, '2000')[['Country Name', '2000']]
top_total_2023 = Total_pop.nlargest(20, '2023')[['Country Name', '2023']]
```

```python
#TotalPopulation in 1960
plt.figure(figsize=(10, 6))
sns.barplot(x='1960', y='Country Name', data=top_total_1960, color='lightblue')

# Adding labels
for i in range(len(top_total_1960)):
    plt.text(top_total_1960['1960'].values[i], i, f"{top_total_1960['1960'].values[i]}", va='center')

plt.title("Top 20 Countries of Total Population (1960)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()

#TotalPopulation in 2000
plt.figure(figsize=(10, 6))
sns.barplot(x='2000', y='Country Name', data=top_total_2000, color='lightblue')

# Adding labels
for i in range(len(top_total_2000)):
    plt.text(top_total_2000['2000'].values[i], i, f"{top_total_2000['2000'].values[i]}", va='center')

plt.title("Top 20 Countries of Total Population (2000)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()

#TotalPopulation in 2023
plt.figure(figsize=(10, 6))
sns.barplot(x='2023', y='Country Name', data=top_total_2023, color='lightblue')

# Adding labels
for i in range(len(top_total_2023)):
    plt.text(top_total_2023['2023'].values[i], i, f"{top_total_2023['2023'].values[i]}", va='center')

plt.title("Top 20 Countries of Total Population (2023)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()
```

### Histogram for Total Population in 1960, 2000, and 2023 

```python
# Histogram for Total Population in 1960
plt.figure(figsize=(10, 6))
sns.histplot(Total_pop['1960'], bins=20, color='blue', kde=True)
plt.title('Distribution of Total Population (1960)')
plt.xlabel('Total Population')
plt.ylabel('Frequency')
plt.show()

# Histogram for Total Population in 2000
plt.figure(figsize=(10, 6))
sns.histplot(Total_pop['2000'], bins=20, color='green', kde=True)
plt.title('Distribution of Total Population (2000)')
plt.xlabel('Total Population')
plt.ylabel('Frequency')
plt.show()

# Histogram for Total Population in 2023
plt.figure(figsize=(10, 6))
sns.histplot(Total_pop['2023'], bins=20, color='red', kde=True)
plt.title('Distribution of Total Population (2023)')
plt.xlabel('Total Population')
plt.ylabel('Frequency')
plt.show()

```
